```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
age = [25,27,29,31,33,35,37,39,41,43,45]
salary= [40000,45000, 50000, 53500, 55000,57500, 60000, 63500, 65000,70000,75000]
```

```python
plt.plot(age,salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary-Age Table")
```

```python
salary2= [45000, 50000, 53500, 55000,57500, 60000, 63500, 65000,70000,75000,77500]
```

```python
plt.plot(age,salary, label="Salary")
plt.plot(age,salary2, label="Salary2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary-Age Table")
plt.legend()
```

### subplot

```python
plt.subplot(2,1,1)  # 2 satır 1 sütünlu plot çizilecek. 1. plot çiziliyor
plt.plot(age, salary, "r")
plt.subplot(2,1,2)   # 2. plot çiziliyor
plt.plot(age, salary2, "b")
plt.tight_layout()
```

```python
plt.subplot(2,2,1)  # 2 satır 2 sütünlu plot çizilecek. 1. plot çiziliyor
plt.plot(age, salary, "r")
plt.subplot(2,2,2)   # 2. plot çiziliyor
plt.plot(age, salary2, "b")
plt.subplot(2,2,3)  # 3 sütünlu plot çizilecek. 1. plot çiziliyor
plt.plot(age, salary, "g")
plt.subplot(2,2,4)   # 4. plot çiziliyor
plt.plot(age, salary2, "y")
plt.tight_layout()
```

```python
plt.plot(salary) # x yok, sadece salary çizdirildi
```
